## 숫자 Image 살펴보기

```python
# 1. package 가져오기
from PIL import Image
import numpy as np
```

```python
# 2. 이미지 가져오기
image_9 = Image.open('01image.png')
image_9
```

```python
# 3. 이미지 array로 변환하기
array_9 = np.array(image_9)
print(array_9)
```

```python
# 4. 이미지 살펴보기
print(array_9.shape) 
```

pixel의 개수 : (28,28) -> 2차원  
흑백 이미지, channel = 1일 경우 생략 가능  
따라서  keras에서 학습 시키기 위해 3차원으로 변환시켜줘야 한다.  
pixel의 개수는 고정한 채 차원만 변환시킴

```python
# 5. 이미지 3차원 reshape
# Dense layer input -> 1차원
# Convolutional layer input -> 3차원

array_9_3d = array_9.reshape(28,28,1)
```

```python
# 6. 이미지 1차원 reshape
array_9_1d = array_9.reshape(28*28)
```

```python
print('1d:\n', array_9_1d)  # 784
```

```python
print('2d:\n', array_9) # 28 * 28
```

```python
print('3d:\n', array_9_3d)
```

```python
# 이미지 시각화
import matplotlib.pyplot as plt
# 1차원은 더이상 이미지가 아니기때문에 시각화 불가능

fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.imshow(array_9, cmap = plt.cm.gray)
ax1.set_title('2D Image')

ax2.imshow(array_9_3d, cmap = plt.cm.gray)
ax2.set_title('3D Image')
plt.show
```

```python
# 7.이미지 resize: 해상도가 너무 높은 경우 사용할 예정
import cv2
image_9_cv = cv2.imread('01image.png')
image_9_cv.resize(10,10)
print(image_9_cv.shape)
```

## 고양이 image

```python
# 고양이 이미지 불러오기
from PIL import Image

cat = Image.open('02image.png')
cat
```

```python
# array 배열
array_cat = np.array(cat)
array_cat
```

```python
# 차원 정보 파악하기
print(array_cat.shape) # 색깔을 가지고 있으므로 channel = 3, (32, 32, 3)이므로 3차원
```

```python
# 3차원
print(array_cat)
```

```python
# 2차원 -> 만들 필요 없음
# array_cat.reshape(32,32) 
#pixel은 동일

array_cat_2d = array_cat[:,:,0] # 하나의 채널만 가져온다(or 평균)
print(array_cat_2d.shape)
print(array_cat_2d)
```

```python
# 1차원 -> Dense layer에 학습가능
array_cat_1d = array_cat.reshape(32*32*3) #pixel은 동일
print(array_cat_1d )
```

## 흑백 이미지로 변환하기 

만약 흐릿하게 불러오는 alpha 값이 있을 경우

cv2.imread을 할 경우  
* 이미지의 alpha 값을 무시함 
* array로 반환

Image.open을 할 경우
*  alpha 값 존재
* np.array() 필요

```python
fruits = Image.open('fruits.jpg')
fruits
```

```python
cv_fruits = cv2.imread('fruits.jpg')
plt.imshow(cv_fruits)
```

```python
cv_fruits.shape
```

cv2.imread를 할 경우 RGB가 반대로 보이게(BGR) 되어 색이 반전됨  
따라서 `plt.show(cv2.cvtColor(객체, cv2.COLOR_BGR2RGB))`를 함

```python
plt.imshow(cv2.cvtColor(cv_fruits, cv2.COLOR_BGR2RGB))
```

```python
cv_fruits_gray = cv2.imread('fruits.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(cv_fruits_gray,cmap = plt.cm.gray)
```
