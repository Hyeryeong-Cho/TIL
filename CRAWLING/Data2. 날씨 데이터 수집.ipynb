```python
import requests 
from bs4 import BeautifulSoup
```

```python
# 기상청 오늘 날씨
response = requests.get("http://www.weather.go.kr/weather/observation/currentweather.jsp")
```

```python
response
```

```python
response.content
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
table = soup.find('table', {'class':'table_develop3'})
```

```python
for tr in table.find_all("tr"):
    print("tr =", tr)
    print("="*50)
```

```python
for tr in table.find_all("tr"):
    tds = tr.find_all("td")
    print("tds =", tds)
    print("="*25)
```

```python
for tr in table.find_all("tr"):
    tds = tr.find_all("td")
    for td in tds:
        print(td)
    print("="*25)
```

```python
for tr in table.find_all("tr"):
    td = tr.find_all("td")
    if len(td) > 0 : # td 태그 안에 들어간 내용이 있을 때
          tds.append(td)
```

```python
for td in tds:
    print("    지점 :", td[0].text)
    print("현재기온 :", td[5].text)
    print("    습도 :", td[10].text)
    print("-"*25)
```

```python
data =[]
for td in tds:
    point = td[0].text
    temperature = td[5].text
    humidity = td[10].text
    data.append([point, temperature, humidity])
```

```python
# with open('weather.csv', 'w') as file: # weather.csv 파일을 쓰기 모드로 열기
#     print("파일 저장합니다.")
#     file.write('point,temperature,humidity\n') # 칼럼 이름 추가
#     for i in data:
#         file.write('{0},{1},{2}\n'.format(i[0],i[1],i[2]))
```

```python
for item in data :
    print("item = ", item)
    row = ",".join(item)
    print("row = ", row)
    print("="*100)
```

```python
with open('weather.csv', 'w') as file: 
    print("파일 저장합니다.") # 파일 저장을 확인하기 위해서
    file.write('point,temperature,humidity\n') # 칼럼 이름 추가
    for item in data:
        row =','.join(item)
        file.write(row+'\n')
```
