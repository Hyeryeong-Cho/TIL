### [모의고사 채점](https://programmers.co.kr/learn/courses/30/lessons/42840/solution_groups?language=python3&type=all)



```python
answers = [1,2,3,4,5,1,2,3,4]
```

```python
n = len(answers)
```

```python
n
```

```python
ans1 = [1,2,3,4,5]*(n-1)
ans2 = [2, 1, 2, 3, 2, 4, 2, 5]*(n-1)
ans3 = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5]*(n-1)
```

```python
ans1[:n]
```

```python
ans2[:n]
```

```python
# 맞힌 개수 확인하기 
cnt1, cnt2, cnt3 =0,0,0

for i in range(0,n):
    if answers[i] == ans1[i]:
        cnt1 += 1
    if answers[i] == ans2[i]:
        cnt2 += 1
    if answers[i] == ans3[i]:
        cnt3 += 1
cnt_lst = [cnt1, cnt2, cnt3]
```

```python
cnt_lst = [9,9,9]
```

```python
# 가장 큰 값을 가진 리스트 위치 찾기
cnt_lst.index(max(cnt_lst)) + 1
```

```python
# list(filter(lambda x+1: cnt_lst[x] == max(cnt_lst)))
```

```python
max_list = [x for x in [0,1,2] if cnt_lst[x] == max(cnt_lst)]
max_list   
```

```python
# 다른 사람 코드
score = [0,0,0]
for idx, answer in enumerate(answers):
        if answer == ans1[idx%len(ans1)]:
            score[0] += 1
        if answer == ans2[idx%len(ans2)]:
            score[1] += 1
        if answer == ans3[idx%len(ans3)]:
            score[2] += 1
```

```python
score
```

```python
result = []
for idx, s in enumerate(score):
    if s == max(score):
        result.append(idx+1)
```

```python
result
```
