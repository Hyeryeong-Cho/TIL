## Simple Keras 모델 생성/학습 - AND Function

```python
# 1. Numpy 가져오기
import numpy as np
print(np.__version__)
```

```python
# 2. And function: 입력/출력 데이터 만들기
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0],[0],[0],[1]])

print(X)
print(y)
```

```python
# 3. Keras 패키지 가져오기
from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import Activation, Dense

print(keras.__version__)
```

```python
# 4. Perceptron 모델 생성
# 모델 생성 및 모델 초기화
model = keras.Sequential()
```

```python
model.add(Dense(input_dim = 2, units = 1)) #feature = 0 or 1(2개)
model.add(Activation('sigmoid'))
model.summary()
```

```python
# 5. Compile - Optimizer, Loss function 설정
model.compile(loss = 'binary_crossentropy', optimizer = 'sgd')
```

```python
# 6. 학습시키기
model.fit(X, y,
          batch_size = 1, # 1개마다 답확인
          epochs = 100) # 100번 반복 => 4개  데이터 * 100 
```

```python
# 7. 모델 테스트하기
test_data = np.array([[0,1], [0,0], [1,1]]) # 0,0,1

print(model.predict(test_data))
print(model.predict_classes(test_data))
```

```python
# optimizer 조정
# ''제외한 sgd 하면 ('sgd': defaluted sgd)

model1 = keras.Sequential()

model1.add(Dense(input_dim = 2, units = 1)) #feature = 0 or 1(2개)
model1.add(Activation('sigmoid'))
model1.summary()

sgd = keras.optimizers.SGD(lr = 0.1)

model1.compile(loss = 'binary_crossentropy', optimizer = sgd)
```

```python
# 모델 학습하기
model1.fit(X, y,
          batch_size = 1, # 1개마다 답확인
          epochs = 100) # 100번 반복 => 4개  데이터 * 100 
```

```python
# 모델 테스트하기
test_data = np.array([[0,1], [0,0], [1,1]]) # 0,0,1

print(model1.predict(test_data))
print(model1.predict_classes(test_data))
```

```python
# 가중치 확인
print('model:', model.get_weights())
print('model1:', model1.get_weights())
```
