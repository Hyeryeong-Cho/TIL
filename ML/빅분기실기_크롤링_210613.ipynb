```python
# 5월 13일 부터 6월 13일까지 데이터 수집
def get_movieDataEachDay(date) :
    base_url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=' + str(date)
    data = requests.get(base_url)
    soup = bs(data.text)
    IstRanking = soup.find(class_ = 'list_ranking')
    
    title = [x.text.strip() for x in IstRanking.find_all('div', class_ = 'tit5')]
    point = [x.text for x in IstRanking.find_all('td', class_ = 'point')]
    rank = range(1, len(point)+1)
    link = ['https://movie.naver.com'+ x.get('href') for x in IstRanking.select('div > a')]

    movieTable = pd.DataFrame({'rank': rank, 'title': title, 'point': point, "link": link})
    movieTable.loc[:, 'date'] = pd.to_datetime(date)
    return movieTable
```

```python
# 날짜별 데이터 프레임 생성
def creatPeriod(start, end): 
    Date = pd.date_range (start=start, end=end, freq = 'd').strftime("%Y%m%d")
    return Date
```

```python
Date = creatPeriod('20210513', '20210613')
Date
```

```python
Movie = pd.DataFrame()

for date in Date : 
    # 일자별 
    df = get_movieDataEachDay(date)
    Movie = pd.concat([Movie, df], axis = 0).reset_index(drop =True)
```
