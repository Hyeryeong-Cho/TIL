```python
# 1. 난수 생성
List =  [7,1,9,5,3,8,4,6,4,2]

#2. 정렬 알고리즘
# 찾는 횟수가 정렬할 위치와 동일
# for 위치 찾는 + for 남아있는 숫자 중 가장 작은 것을 찾는 

n = len(List)

for i in range(0, n-1): # 마지막은 한개만 남아서 n-1번만 하면되어서, 즉 index는 n-2까지!
    # 정렬되지 않은 숫자 중 가장 작은 숫자 선택
    Min = i # 정렬할 위치 찾기 찾기
    for j in range(i+1, n):
        
        if List[j] < List[Min] :
            Min = j
        if Min != i :
            temp = List[i] 
            List[i] = List[Min]
            List[Min] = temp
        
    print(List)
```

## Count sort 구현
1. 1~7사이의 무작위 값 20개인 listA
2. listA의 최대값과 최소값을 찾아 C리스트 생성
    (0으로 채워져있는 길이가 ndls리스트)
3. listA의 각 index를 확인하면서 C리스트의 값을 업데이트
4. C리스트의 업데이트를 끝내면, listB를 생성하고 채우시오.

```python
import random

length = 10

num =0
listA = []

while num < length :
    listA.append(random.randint(1,7))
    num = num +1
    


Max = -999
Min = 999
for i in range(0,len(listA)):
    if Max < listA[i]:
        Max = listA[i]
print('최대값은: ', Max)

for i in range(0,len(listA)):
    if Min > listA[i]:
        Min = listA[i]
print('최소값은: ', Min)
        
```

```python
num = 0
listC =[]        


while num < (Max-Min+1) :
    listC.append(0)
    num = num +1
print(listA,listC )

num = 0
while num < len(listA) :
    listC[listA[num]-Min] =(listC[listA[num]-Min]+1) 
    num= num+1
    print(listA,listC )
input("카운트 리스트 완성")

num=0
```

```python
# 최소값을 증가시키는 버전
IdxC = 0
listB = []
while IdxC < len(listC):
    if listC[IdxC] != 0 :
        listB.append(IdxC+Min)
        listC[IdxC] = listC[IdxC]-1
        print(listA, listC, listB)
    else: IdxC += 1
        
```
