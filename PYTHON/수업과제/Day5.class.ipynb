```python
#원 클래스 만들기
class Circle :
    def __init__(self, radius=1.0):
        self.__radius = radius
    def setRadius(self, r):
        self.__radius = r
    def getRadius(self):
        return self.__radius
    def calcArea(self):
        area = 3.14*self.__radius*self.__radius
        return area
    def calcCircum(self):
        circumference = 2.0*3.14*self.__radius
        return circumference
```

```python
c = Circle()
```

```python
c.getRadius()
```

```python
c.setRadius(5)
```

```python
c.getRadius()
```

```python
c.calcArea()
```

```python
print('원의 반지름: ', c.getRadius())
print('원의 넓이: ', c.calcArea())
print('원의 둘레: ', c.calcCircum())
```

```python
# 은행계좌를 만들어보자
class BankAccount:
    def __init__(self):
        self.__balance = 0
    def withdraw(self, amount):
        self.__balance -= amount
        print('통장에서', amount,'이 출금되었음')
    def deposit(self, amount):
        self.__balance += amount
        print('통장에', amount,'이 입금되었음')
    def bankBalance(self):
        return self.__balance
```

### Class method의 호출
- instance와 class함수 이름이 같을 경우

```python
class miner:
    totalMineral = 0
    def __init__(self,mineral):
        self.mineral = mineral
        miner.totalMineral +=mineral
    def miningMineral(self):
        self.mineral += 10
        miner.totalMineral += 10
    @classmethod
    def printMineral(cls):
        print(cls.totalMineral)
```

```python
SCV = miner(50)
probe = miner(100)
miner.printMineral()
```

```python
class miner:
    totalMineral = 0
    def __init__(self,mineral):
        self.mineral = mineral
        miner.totalMineral +=mineral

    @classmethod
    def printMineral(cls):
        print(cls.totalMineral)
        
```

```python
SCV = miner(50)
probe = miner(100)
miner.printMineral() #classmethod가 나옴
```
