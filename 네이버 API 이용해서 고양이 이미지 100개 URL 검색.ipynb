# Open API로 데이터 수집하기

```python
def get_api_result(keyword, display, start):
    url = "https://openapi.naver.com/v1/search/image?query="\
            + keyword + "&display=" + str(display) + '&start=' + str(start)
    result = requests.get(url, headers = api_id_pw)
    return result.json()
```

```python
get_api_result("강아지", 100, 1)
```

### 1000개씩 이미지 가져오는 함수 

```python
def call_and_print(keyword, total_page =10) : # 만 개일 때는 100
    link_lst = [] # 저장할 링크 리스트
    for page in range(1, total_page+1): # 1~10
        display = 100
        start = ((page-1))*display+1 # 전 페이지 *100 +1 부터 시작
        json_obj = get_api_result(keyword,display, start)
        # link 만 저장
        for item in  json_obj['items']:
            link_lst.append(item['link'])

    return link_lst
```

```python
keyword = '고양이'
link_cat = call_and_print(keyword)
```

```python
keyword = '강아지'
link_cat = call_and_print(keyword)
```

```python
keywords = ['고양이','강아지']
link_all = []
for keyword in keywords:
    lst=[]
    print(keyword+'의 링크를 수집중입니다!')
    lst = call_and_print(keyword)
    link_all+=lst
```
