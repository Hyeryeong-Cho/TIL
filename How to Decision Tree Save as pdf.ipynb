```python
# Îç∞Ïù¥ÌÑ∞ Î∂ÑÎ¶¨
# Ïî®Îìú ÏÑ§Ï†ï
x_train, x_test, y_train, y_test = train_test_split(youtube["title"], youtube["view_num2"], random_state=101)
```

## Îã®Ïñ¥Î°ú Î∂ÑÎ¶¨ÌïòÍ∏∞

```python
### Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨
from konlpy.tag import Twitter
twitter = Twitter()
```

```python
from konlpy.tag import Twitter
twitter = Twitter()

def getNounAdject(text):
    # Ïñ¥Í∞ÑÏ∂îÏ∂ú
    tagged = twitter.pos(text, stem=True)
    stems = [tagged[i][0] for i in range(len(tagged)) if (tagged[i][1]=="Noun") | (tagged[i][1] =="Adjective")]
    print("stems = ", stems)
    print("="*100)
    return stems
```

## Îã®Ïñ¥Î•º Ïà´ÏûêÎ°ú Î≥ÄÌôò TF-IDF

```python
## TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# ÎëêÎ≤à Ïù¥ÏÉÅ ÎÇòÏò® Îã®Ïñ¥Îßå Ïà´ÏûêÎ°ú Î≥ÄÌôò 
# 
vectorizer = TfidfVectorizer(min_df = 2, tokenizer= getNounAdject)
```

```python
x_train_vector = vectorizer.fit_transform(x_train)
```

```python
# x train
df_tfidf1 = pd.DataFrame(x_train_vector.A, columns=vectorizer.get_feature_names())
df_tfidf1
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier()
```

```python
dt_clf = dt_clf.fit(x_train_vector, y_train)
```

### test data ÏÑ±Îä• ÌèâÍ∞Ä

```python
x_test_vector = vectorizer.transform(x_test)
```

```python
df_prediction = dt_clf.predict(x_test_vector)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, df_prediction)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, df_prediction)
print("Ï†ïÌôïÎèÑ:{0}%".format(accuracy))
```
