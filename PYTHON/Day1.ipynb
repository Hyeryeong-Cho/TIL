## 조건반복문 -while
- 조건이 만족하는 동안 실행문장을 반복한다.

```python
# 같은 문장 5번 반복
cnt=0 # cnt=count

while cnt <5 : #0부터 시작하기때문에 5번 반복하려면 <5임!
    print(cnt, "파이썬은 재밌습니다!")
    cnt=cnt+1
    print(cnt)
print('cnt가 5이므로 조건 반복 끝!')
```
