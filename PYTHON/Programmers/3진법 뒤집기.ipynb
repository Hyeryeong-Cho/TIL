## 3진법 뒤집기


```python
# 3진법 Ternary 도출하기
def Ternary(num, div):
    result = []
    q, r = num, div
    while q > 0:
        q, r = divmod(q, div)
        if r > 10:
            _remain = 'ABCDEF'
            r = _remain[r-10]
        result.append(str(r))
    print(''.join(result[::-1]))
```

```python
q, r = 7,3
q, r = divmod(7,3)

result = [str(q),str(r)] # join은 str만 가능

''.join(result)
```
