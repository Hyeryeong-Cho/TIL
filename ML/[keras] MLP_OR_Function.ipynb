### OR Function

```python
# 패키지 불러오기
import numpy as np
```

```python
# OR 함수 
X = np.array([[0,0],[0,1],[1,0],[1,1]])

y = np.array([[0],[1],[1],[1]])

print(X)
print(y)
```

```python
# deep learning
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Activation, Dense
```

```python
# 모델 생성하기
model = keras.Sequential()

model.add(Dense(input_dim = 2, units = 1)) #feature = 0 or 1(2개)
model.add(Activation('sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'sgd')

model.fit(X,y, batch_size = 1, epochs = 100)
```

```python
# 모델 생성하기
model1 = keras.Sequential()

model1.add(Dense(input_dim = 2, units = 1)) #feature = 0 or 1(2개)
model1.add(Activation('sigmoid'))

model1.compile(loss = 'binary_crossentropy', optimizer = 'adam')

model1.fit(X,y, batch_size = 1, epochs = 100)
```

```python
test_data = np.array([[0,1], [0,0], [1,1]]) # 1,0,1
print('sgd:\n', model.predict(test_data))
print('sgd:\n',model.predict_classes(test_data))
print('adam:\n', model1.predict(test_data))
print('adam:\n',model1.predict_classes(test_data))
```

```python
# 모델 생성하기
model2 = keras.Sequential()

model2.add(Dense(input_dim = 2, units = 1)) #feature = 0 or 1(2개)
model2.add(Activation('sigmoid'))

sgd= keras.optimizers.SGD(lr = 9)

model2.compile(loss = 'binary_crossentropy', optimizer = sgd)

model2.fit(X,y, batch_size = 1, epochs = 100)
```

```python
print('sgd_lr_0.1:\n', model2.predict(test_data))
print('sgd_lr_0.1:\n',model2.predict_classes(test_data))
```

* Adam과 SGD의 차이
* learning rate를 높게 해줘야 예측률이 향상
* learning rate의 범위는 0이상
